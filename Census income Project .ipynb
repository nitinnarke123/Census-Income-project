# Classification model

```python
#Logistic 
#Decission Tree
#Random forest
```

```python
#splittnig the data into dependent and independent
x=data.iloc[:,:-1]
```

```python
x
```

```python
y=data["marital-status"]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)
```

```python
#implementing Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lo_Model=LogisticRegression()
```

```python
lo_Model.fit(x_train,y_train)
```

```python
lo_pred=lo_Model.predict(x_test)
```

```python
lo_pred
```

```python
y_test
```

```python
lo_accuracy=accuracy_score(lo_pred,y_test)*100
```

```python
lo_accuracy
```

```python
cm=confusion_matrix(lo_pred,y_test)
```

```python
sns.heatmap(cm,annot=True,fmt="g")
```

```python
#implementation of decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_Model=DecisionTreeClassifier()
```

```python
dt_Model.fit(x_train,y_train)
```

```python
dt_pred=dt_Model.predict(x_test)
```

```python
dt_pred
```

```python
y_test
```

```python
dt_accuracy=accuracy_score(dt_pred,y_test)*100
```

```python
dt_accuracy
```

```python
cm=confusion_matrix(dt_pred,y_test)
sns.heatmap(cm,annot=True,fmt="g")
```

```python
#implementation of  Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_Model=RandomForestClassifier()
```

```python
rf_Model.fit(x_train,y_train)
```

```python
rf_pred=rf_Model.predict(x_test)
```

```python
rf_pred
```

```python
y_test
```

```python
rf_accuracy=accuracy_score(rf_pred,y_test)*100
```

```python
rf_accuracy
```

```python
cm=confusion_matrix(rf_pred,y_test)
sns.heatmap(cm,annot=True,fmt="g")
```

```python
print("Accuracy of Logistic Regression :",lo_accuracy)
print("Accuracy ofDecision Tree :",dt_accuracy)
print("Accuracy of Random Forest :",rf_accuracy)
```
